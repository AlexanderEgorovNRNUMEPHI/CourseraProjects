```python
import math as math
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

```python
def func(x):
    return (math.sin(x/5.))*math.exp(x/10.)+5.*math.exp(-(x/2.))
```

```python
myfunc=[]
arg=[]
for i in range(1,31,1):
    func(i)
    myfunc.append(func(i))
    arg.append(i)
    
    
```

```python
plt.figure(figsize=(7, 7), dpi=80)
plt.plot(arg, myfunc)
plt.xlim(0, 30)
plt.xticks(np.arange(0, 30, 2),  color='black', size=14)
plt.ylim(-12, 4)
plt.yticks(np.arange(-12, 4, 0.5),  color='black', size=14)
plt.show()
```

```python
arrmyfunc=np.array(myfunc)
```

```python

optimize.minimize(func,2, method='BFGS')
```

```python
optimize.differential_evolution(func,[(1,30)])
```
